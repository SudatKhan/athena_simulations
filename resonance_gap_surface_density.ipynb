```python
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
import matplotlib.pyplot as plt
sys.path.insert(0, '/home/sudat/athena/vis/python')
import athena_read
```

```python
'''Reading data from HDF5 files'''
sum_dens = 0
for i in range(3500,4000):
    data = athena_read.athdf(f'/home/sudat/resonance_two_planet_simulations/disk.out1.0{i}.athdf')
    density = data.get('dens')[0]
    sum_dens += density

sum_dens_avg = sum_dens/len(range(3500,4000))
```

```python
sum_dens_onesim = 0
for i in range(3500,4000):
    data_onesim = athena_read.athdf(f'/home/sudat/one_planet_simulations/disk.out1.0{i}.athdf')
    density_onesim = data_onesim.get('dens')[0]
    sum_dens_onesim += density_onesim

sum_dens_avg_onesim = sum_dens_onesim/len(range(3500,4000))
```

```python
r = data.get('x1v')
phi = data.get('x2v')
plt.plot(r, np.average(sum_dens_avg, axis = 0), label = 'One-Planet Simulation')
plt.plot(r, np.average(sum_dens_avg_onesim, axis = 0), label= 'Two-Planet Simulation')
plt.yscale('log')
plt.ylabel(r'$\langle {\Sigma}/{\Sigma_{0}}\rangle_{\phi}$')
plt.xlabel(r'$R \,[AU]$')
plt.legend()
```
