```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '/home/sudat/athena/vis/python')
import athena_read
```

```python
'''Reading data from HDF5 files'''
data1 = athena_read.athdf('disk.out1.00583.athdf')
data2 = athena_read.athdf('disk.out3.00583.athdf')
density = data1.get('dens')[0]
r = data1.get('x1f')
phi = data1.get('x2f')
g_r = data2.get('user_out_var0')[0]
g_phi = data2.get('user_out_var1')[0]
```

```python
'''Conversion from Polar -> Cartesian'''
x = r[None,:] * np.cos(phi[:,None])
y = r[None,:] * np.sin(phi[:, None])
```

```python
'''Plot 2-D surface density mesh plot'''
plt.pcolormesh(x, y, np.log(density), cmap='inferno')
```

```python
'''Plot gravitational acceleration mesh plot in r-direction'''
plt.pcolormesh(x, y, g_r, cmap = 'RdBu', vmin=-0.05, vmax= 0.05)
plt.colorbar()
```

```python
'''Plot gravitational acceleration mesh plot in phi-direction'''
plt.pcolormesh(x,y, g_phi, cmap = 'RdBu', vmin=-0.05, vmax= 0.05)
plt.colorbar()
```
